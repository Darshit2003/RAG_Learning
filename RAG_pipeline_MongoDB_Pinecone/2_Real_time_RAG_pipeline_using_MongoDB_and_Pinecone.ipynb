```python
from pinecone import Pinecone

pc = Pinecone(api_key=PINECONE_API_KEY)
index = pc.Index("mongomovie")
```

```python
def get_details(query,similar_result=3):
  embedding = embedding_model.encode(query).tolist()
  result = index.query(
      vector=embedding,
      top_k=similar_result,
      include_values=True
  )
  return result
```

```python
query = "Which is the best horror movie to watch and why?"
```

```python
result = get_details(query)
```

```python
print(result)
```

#To LLM i cannnot send this vectors i need to send text which i will get from MongoDB only

```python
from bson.objectid import ObjectId

my_list = []
for match in result['matches']:
    value = match['id']  # should be a string ObjectId
    doc = collection.find_one({"_id": ObjectId(value)})  # fetch full record
    my_list.append(doc)
```

```python
print(my_list)
```

```python
combined_information = ""

for doc in my_list:
    title = doc.get("title", "N/A")
    fullplot = doc.get("fullplot", "N/A")

    combined_information += f"📽️ Title: {title}\n📝 fullplot: {fullplot}\n\n"
```

```python
print(combined_information)
```

```python
import os
from google.colab import userdata
from langchain_cohere import ChatCohere

# Set Cohere API key
COHERE_API_KEY = userdata.get('COHERE_API_KEY')
os.environ["COHERE_API_KEY"] = COHERE_API_KEY

def load_model(model_name="command-r-plus"):
    llm = ChatCohere(model=model_name, temperature=0.7, max_tokens=512)
    return llm

llm = load_model("command-r-plus")

prompt_template = f"""
You are a knowledgeable movie assistant. Use the following context from the database
to answer the user's question. Do not use outside knowledge. If the answer is not
in the context, just say "I don't know based on the available data."

Context:
{{context}}

Question:
{{question}}

Answer:
"""
user_question = "Which is the most horror movie?"

final_prompt = prompt_template.format(
    context=combined_information,
    question=user_question
)



# Run your prompt
response = llm.invoke(final_prompt)
print(response.content)
```
